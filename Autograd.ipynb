```python
# Automatic differentiation
import autograd.numpy as np  # Thinly-wrapped numpy
from autograd import grad

# Plotting
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}

# Interaction
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
```

```python
def f(x):
    return np.sin(x)**2
```

```python
def derivative(x):
    return grad(f)(x)
```

```python
XRange = np.arange(-5,5,0.1)

@interact(c=(-5.0,5.0,0.1) )
def plotFunctions(c):
    plt.xlim([-5,5])
    plt.ylim([-1,2])
    
    # f(x)
    plt.plot(XRange, f(XRange))
    
    # c 
    plt.scatter(c, f(c))
    
    # tangent line
    plt.plot(XRange, derivative(c)*XRange + (f(c)-derivative(c)*c))
    plt.show()
    
    return "Derivative is " + str(derivative(c))
    
```
